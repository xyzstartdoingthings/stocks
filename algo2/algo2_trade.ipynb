```python
import pickle
import os
from pathlib import Path
from algo_2 import *
```

```python
with open('final_pick','rb') as f:
    final_pick = pickle.load(f)
final_pick
```

```python
# List all files in the directory that start with "output "
directory = os.getcwd()
pickle_files = [f for f in os.listdir(directory) if f.startswith("output finer")]

# Initialize a dictionary to store the contents of each pickle file
pickle_contents = []

# Loop through each file, read its content, and store it in the dictionary
for file_name in pickle_files:
    file_path = os.path.join(directory, file_name)
    with open(file_path, 'rb') as file:
        pickle_contents.extend(pickle.load(file))

# pickle_contents.keys()
pickle_contents.sort(key=lambda x: x[0][1], reverse=True)
```

```python
today_dict={}
for i in pickle_contents:
    for j in range(len(i)):
        if i[j][0]["ticker"] in final_pick:
            stock, short_mark, long_mark, peak_dict, trough_dict, stoploss, stopgain = algo2_trade(**i[j][0])
            if (stock.loc[len(stock)-1, "order"] != 0) or long_mark!=0 or short_mark!= 0:
                today_dict[i[j][0]["ticker"]] = {}
                today_dict[i[j][0]["ticker"]]["date"] = stock.loc[len(stock)-1, "date"]
                today_dict[i[j][0]["ticker"]]["order"] = stock.loc[len(stock)-1, "order"]
                today_dict[i[j][0]["ticker"]]["short_mark"] = short_mark
                today_dict[i[j][0]["ticker"]]["long_mark"] = long_mark
                today_dict[i[j][0]["ticker"]]["peak_dict"] = peak_dict
                today_dict[i[j][0]["ticker"]]["trough_dict"] = trough_dict
                today_dict[i[j][0]["ticker"]]["stoploss"] = stoploss
                today_dict[i[j][0]["ticker"]]["stopgain"] = stopgain
                today_dict[i[j][0]["ticker"]]["variable"] = i[j][0]
                continue

            
```
